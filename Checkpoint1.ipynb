```python
# Tratamento dos dados
import numpy as np
import pandas as pd

# Modelos de Machine Learning
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split

#Importando os dados
df = pd.read_csv('user_navigation_data.csv', sep=',', header=None)
```

```python
df.columns = ["id","user_id","session_id","timestamp","time_spent", "page_views","actions", "compra"]
```

```python
df.head(50)
```

```python
df['actions'] = df['actions'].replace({'click': 0, 'view':1, })
df['compra'] = df['compra'].replace({'comprou': 0, 'nao_comprou':1, })

#df['timestamp'] = df['timestamp'].str.replace('-','')
#df['timestamp'] = df['timestamp'].str.replace(':','')
#df['timestamp'] = df['timestamp'].str.replace(' ','')

#df['timestamp']=df['timestamp'].astype(float)
```

```python
x = df.drop(columns = { 'actions' , 'timestamp', 'session_id', 'id', 'user_id'})
y = df['actions']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
redeneural = MLPClassifier(verbose = True, max_iter = 500, tol = 0.000001, activation = 'relu', #hidden-layer_sizes=(30,30,30), 
                           learning_rate_init= 0.000001)
```

```python
redeneural.fit(x_train, y_train)
```

```python
redeneural.score(x_test, y_test)
```
